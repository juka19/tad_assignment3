```python
import pandas as pd

df = pd.read_csv("data/data_28_11_sponsors.csv")
for index,items in enumerate(df.summary):
    if type(items) != str:
        print(type(items), index)
df.drop(681, axis=0, inplace=True)
df.summary = list(map(lambda x: x.lower() ,df.summary))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from stop_words import get_stop_words 
from sklearn.decomposition import NMF




vectorizer = TfidfVectorizer(stop_words=get_stop_words('en'))
dfm_summary = vectorizer.fit_transform(df.iloc[2]["summary"].split("."))
N_topics = 4

#second_summary = dfm_summary[2,:].shape 
nmf = NMF(N_topics)
nmf.fit_transform(dfm_summary)


#t_titles = plot_top_words(nmf, vectorizer.get_feature_names_out(df_old_parties.text), words_for_topic, "Topics in existing parties in " + str(year), )

#n_top_words = 6

# Create a DataFrame: components_df
components_df = pd.DataFrame(nmf.components_, columns=vectorizer.get_feature_names())
components_df
  

def print_top_words(model, n_top_words, feature_names= vectorizer.get_feature_names_out(df.summary)):
    total_words = []
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
        top_features = [feature_names[i] for i in top_features_ind]
        #print("Topic: ", top_features)
        total_words.extend(top_features)
    return total_words

topics = []
for summary in df.summary:
    dfm_summary = vectorizer.fit_transform(summary.split("."))
    nmf.fit_transform(dfm_summary)

    #components_df = pd.DataFrame(nmf.components_, columns=vectorizer.get_feature_names())
    topics.append(print_top_words(nmf, 3, feature_names= vectorizer.get_feature_names_out(df.summary)))

    
df["topics"] = topics
df
```

```python
import csv

N_DIMS = 100
f = open(f'../glove/glove.6B.{N_DIMS}d.txt')

word_matrix = pd.read_table(
    f, sep=" ", index_col=0, 
    header=None, quoting=csv.QUOTE_NONE
)
word_matrix[word_matrix.index == "hezbollah"]
```

```python
val = word_matrix[word_matrix.index == "sec"].values
val.shape
```

```python
import numpy as np

def from_list_to_matrix(lst):
    mtrx = np.zeros((100,12))
    for i,word in enumerate(lst):
        if word in word_matrix.index:
            mtrx[:, i] = word_matrix.loc[word].values
        else:
            mtrx[:, i] = np.zeros(100)
    return mtrx

df["topics_matrix"] = df["topics"].apply(from_list_to_matrix)
```

```python
df["topics_matrix"][0].shape
```

```python
from sklearn.model_selection import train_test_split

def constr(x):
    if x>0.65:
        return 2
    elif 0.65 >= x >= 0.35:
        return 1
    else:
        return 0


X = df["topics_matrix"].values
y = df["cosponsor_D_perc"].values
y = list((map(constr, y)))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=40)
print(X_train.shape); print(X_test.shape)
y_train
```

```python
X_train[0].flatten().shape
```

```python
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from tensorflow import convert_to_tensor


y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
print(y_train[0])
y_train = [y.flatten() for y in y_train]
y_test = [y.flatten() for y in y_test]

y_train = [y.reshape(1,3) for y in y_train]
y_test = [y.reshape(1,3) for y in y_test]

# Dense(64) is a fully-connected layer with 64 hidden units.
# in the first layer, you must specify the expected input data shape:
# here, 20-dimensional vectors.

X_train = [X.flatten() for X in X_train]
X_train = [X.reshape(1,1200) for X in X_train]
X_test = [X.flatten() for X in X_test]
X_test = [X.reshape(1,1200) for X in X_test]

#y_train = [convert_to_tensor(y) for y in y_train]
#y_test = [convert_to_tensor(y) for y in y_test]
```

```python
X_train[0].reshape(1,1200).shape
```

```python
model = Sequential()

model.add(Dense(1200, activation='tanh', input_dim=1))
model.add(Dropout(0.5))
model.add(Dense(64, activation='tanh'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='softmax'))
model.add(Dense(3, activation='softmax'))

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)

model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,epochs=100, batch_size=7)
print(len(X_train), len(y_train))
```
