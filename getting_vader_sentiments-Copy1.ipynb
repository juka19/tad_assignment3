```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import requests
import io
```

```python
# Downloading the csv file from your GitHub account

url = "https://raw.githubusercontent.com/juka19/tad_assignment3/main/data/data_11_28.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))

df
```

```python
df1 = df.dropna(subset=['summary'])
```

```python
analyzer = SentimentIntensityAnalyzer()

results = [analyzer.polarity_scores(x) for x in df1["summary"]] 

sentiment = pd.DataFrame.from_dict(results)

sentiment_df = pd.concat([df1,sentiment], axis=1) 

sentiment_df
```
